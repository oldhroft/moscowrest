```python
import pandas as pd
import numpy as np
```

```python
import json
```

```python
with open('../data/stg/mosdata_dataset3304.json', 'r') as file:
    data = json.load(file)
```

```python
def process(item: dict):
    
    result = {'global_id': item['global_id'], 'Number': item['Number']}
    result.update(item['Cells'])
    result['x_coord'] = result['geoData']['coordinates'][0]
    result['y_coord'] = result['geoData']['coordinates'][1]
    result['PublicPhone'] = list(map(lambda x: list(x.values())[0], result['PublicPhone']))
    del result['geoData']
    return result
```

```python
data_expanded = pd.DataFrame(list(map(process, data)))
```
